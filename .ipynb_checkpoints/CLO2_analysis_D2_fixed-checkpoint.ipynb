## 8. Hypothesis testing

### 8.1 Pearson correlation (numeric IV vs DV)

```python
pearson_corr, p_value = stats.pearsonr(df["Avg_Daily_Usage_Hours"], df["Addicted_Score"])

print("Pearson Correlation:", pearson_corr)
print("P-value:", p_value)
```

### 8.2 Spearman correlation (numeric IV vs DV)

```python
spearman_corr, p_value = stats.spearmanr(df["Sleep_Hours_Per_Night"], df["Addicted_Score"])

print("Spearman Correlation:", spearman_corr)
print("P-value:", p_value)
```

### 8.3 Chi-square test (categorical vs categorical)

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(df["Gender"], df["Relationship_Status"])

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square:", chi2)
print("P-value:", p)
```

## 9. One-sample t-test (Is the random sample representative of the dataset

```python
population_mean = df["Addicted_Score"].mean()
t_stat, p_value = stats.ttest_1samp(random_sample["Addicted_Score"], population_mean)
print("Population mean (Addicted_Score):", population_mean)
print("Random sample mean:", random_sample["Addicted_Score"].mean())
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

# 10. Deliverable 2 – Machine Learning Models

## 10.1 Data preprocessing for machine learning

```python
dataset = df.copy()

# Remove duplicates
dataset = dataset.drop_duplicates()

# Fill missing values
numeric_columns = dataset.select_dtypes(include=np.number).columns
categorical_columns = dataset.select_dtypes(exclude=np.number).columns

for col in numeric_columns:
    dataset[col] = dataset[col].fillna(dataset[col].mean())

for col in categorical_columns:
    dataset[col] = dataset[col].fillna(dataset[col].mode()[0])

print("Dataset shape after preprocessing:", dataset.shape)
print("\nMissing values after preprocessing:")
print(dataset.isnull().sum())
```

## 10.2 Preparing the data for regression

```python
# Dependent variable (target)
y_reg = dataset['Addicted_Score']

# Independent variables (drop target and ID)
X_reg = dataset.drop(columns=['Addicted_Score', 'Student_ID'])

# Encode categorical independent variables
X_reg = pd.get_dummies(X_reg, drop_first=True)

print("Regression features shape:", X_reg.shape)
print("Regression target shape:", y_reg.shape)
X_reg.head()
```

## 10.3 Splitting the dataset for regression

```python
from sklearn.model_selection import train_test_split

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

print("X_train_reg:", X_train_reg.shape)
print("X_test_reg:", X_test_reg.shape)
print("y_train_reg:", y_train_reg.shape)
print("y_test_reg:", y_test_reg.shape)
```

## 10.4 Feature scaling for regression

```python
from sklearn.preprocessing import StandardScaler

sc_reg = StandardScaler()
X_train_reg_scaled = sc_reg.fit_transform(X_train_reg)
X_test_reg_scaled = sc_reg.transform(X_test_reg)
```

## 11. Regression models

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score

regression_results = []

# Multiple Linear Regression
multiple_linear = LinearRegression()
multiple_linear.fit(X_train_reg_scaled, y_train_reg)
y_pred_multiple = multiple_linear.predict(X_test_reg_scaled)
regression_results.append(['Multiple Linear Regression',
                           mean_squared_error(y_test_reg, y_pred_multiple),
                           r2_score(y_test_reg, y_pred_multiple)])

# Polynomial Regression (Degree 2)
polynomial_model = Pipeline([
    ('poly', PolynomialFeatures(degree=2)),
    ('linear', LinearRegression())
])
polynomial_model.fit(X_train_reg_scaled, y_train_reg)
y_pred_polynomial = polynomial_model.predict(X_test_reg_scaled)
regression_results.append(['Polynomial Regression',
                           mean_squared_error(y_test_reg, y_pred_polynomial),
                           r2_score(y_test_reg, y_pred_polynomial)])

# Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train_reg_scaled, y_train_reg)
y_pred_ridge = ridge_model.predict(X_test_reg_scaled)
regression_results.append(['Ridge Regression',
                           mean_squared_error(y_test_reg, y_pred_ridge),
                           r2_score(y_test_reg, y_pred_ridge)])

# Lasso Regression
lasso_model = Lasso(alpha=0.01)
lasso_model.fit(X_train_reg_scaled, y_train_reg)
y_pred_lasso = lasso_model.predict(X_test_reg_scaled)
regression_results.append(['Lasso Regression',
                           mean_squared_error(y_test_reg, y_pred_lasso),
                           r2_score(y_test_reg, y_pred_lasso)])

# Elastic Net Regression
elastic_model = ElasticNet(alpha=0.01, l1_ratio=0.5)
elastic_model.fit(X_train_reg_scaled, y_train_reg)
y_pred_elastic = elastic_model.predict(X_test_reg_scaled)
regression_results.append(['Elastic Net Regression',
                           mean_squared_error(y_test_reg, y_pred_elastic),
                           r2_score(y_test_reg, y_pred_elastic)])

regression_results_df = pd.DataFrame(regression_results, columns=['Model', 'MSE', 'R_squared'])
regression_results_df = regression_results_df.sort_values(by='R_squared', ascending=False)
print(regression_results_df)
```

### Regression plots for all 5 models

```python
def plot_regression_comparison(y_true, y_pred, title):
    plt.figure(figsize=(6, 4))
    plt.scatter(y_true, y_pred, alpha=0.7)
    plt.plot([y_true.min(), y_true.max()],
             [y_true.min(), y_true.max()],
             color='red', linestyle='--')
    plt.title(title)
    plt.xlabel('Observed Addicted Score')
    plt.ylabel('Predicted Addicted Score')
    plt.show()

plot_regression_comparison(y_test_reg, y_pred_multiple, 'Multiple Linear Regression')
plot_regression_comparison(y_test_reg, y_pred_polynomial, 'Polynomial Regression')
plot_regression_comparison(y_test_reg, y_pred_ridge, 'Ridge Regression')
plot_regression_comparison(y_test_reg, y_pred_lasso, 'Lasso Regression')
plot_regression_comparison(y_test_reg, y_pred_elastic, 'Elastic Net Regression')
```

## 12. Classification models

### Defining the target variable as a categorical value

```python
dataset['Addiction_Level'] = pd.cut(
    dataset['Addicted_Score'],
    bins=[0, 4, 7, 10],
    labels=['Low', 'Medium', 'High'],
    include_lowest=True
)

print(dataset['Addiction_Level'].value_counts())
```

### Preparing classification features and target

```python
X_clf = dataset.drop(columns=['Addicted_Score', 'Addiction_Level', 'Student_ID']).copy()
X_clf = pd.get_dummies(X_clf, drop_first=True)

y_clf = dataset['Addiction_Level']

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_clf_encoded = label_encoder.fit_transform(y_clf)

print("Classification features shape:", X_clf.shape)
print("Classification target shape:", y_clf_encoded.shape)
```

### Splitting and scaling the classification dataset

```python
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf_encoded, test_size=0.2, random_state=42, stratify=y_clf_encoded
)

sc_clf = StandardScaler()
X_train_clf_scaled = sc_clf.fit_transform(X_train_clf)
X_test_clf_scaled = sc_clf.transform(X_test_clf)
```

### Building and comparing classification models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

classification_results = []

classification_models = {
    'Logistic Regression': LogisticRegression(max_iter=2000),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'SVM': SVC(kernel='rbf', random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42)
}

for model_name, model in classification_models.items():
    if model_name in ['Decision Tree', 'Random Forest']:
        model.fit(X_train_clf, y_train_clf)
        y_pred = model.predict(X_test_clf)
    else:
        model.fit(X_train_clf_scaled, y_train_clf)
        y_pred = model.predict(X_test_clf_scaled)

    accuracy = accuracy_score(y_test_clf, y_pred)
    precision = precision_score(y_test_clf, y_pred, average='weighted')
    recall = recall_score(y_test_clf, y_pred, average='weighted')
    f1 = f1_score(y_test_clf, y_pred, average='weighted')

    classification_results.append([model_name, accuracy, precision, recall, f1])

    print("\n", model_name)
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1-score:", f1)
    print(classification_report(y_test_clf, y_pred, target_names=label_encoder.classes_))

classification_results_df = pd.DataFrame(
    classification_results,
    columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1_score']
).sort_values(by='F1_score', ascending=False)

print("\nComparison of Classification Models")
print(classification_results_df)
```

### Selecting the best-fit classifier

```python
best_classifier_name = classification_results_df.iloc[0]['Model']
print("Best-fit classifier:", best_classifier_name)

best_classifier = classification_models[best_classifier_name]

if best_classifier_name in ['Decision Tree', 'Random Forest']:
    best_predicted = best_classifier.predict(X_test_clf)
else:
    best_predicted = best_classifier.predict(X_test_clf_scaled)

predicted_labels = label_encoder.inverse_transform(best_predicted)
print("\nNumber of elements in each predicted category:")
print(pd.Series(predicted_labels).value_counts())
```

## 12. Cluster Analysis

### Choosing the clustering features
For clustering, the selected features are **Avg_Daily_Usage_Hours** and **Sleep_Hours_Per_Night** because they represent student behaviour patterns more meaningfully.

```python
X_cluster = dataset[['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night']].values
print(X_cluster[:10])
```

### Using the Elbow Method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_cluster)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Training the K-Means model on the dataset
Based on the elbow method, **3 clusters** are used.

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X_cluster)

print(y_kmeans)
```

### Visualising the clusters

```python
plt.scatter(X_cluster[y_kmeans == 0, 0], X_cluster[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X_cluster[y_kmeans == 1, 0], X_cluster[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X_cluster[y_kmeans == 2, 0], X_cluster[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=250, c='yellow', label='Centroids')

plt.title('Clusters of Students')
plt.xlabel('Mental_Health_Score')
plt.ylabel('Conflicts_Over_Social_Media')
plt.legend()
plt.show()
```

### Strategy formulation

**Suggested interpretation and strategy**
- **Cluster 1:** Students with low mental health scores and high conflicts over social media
 
- **Cluster 2:** Students with high mental health scores and low conflicts over social medi

- **Cluster 3:** Students with moderate mental health scores and moderate conflicts over social media
  

### Hierarchical clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(X_cluster, method='ward')

plt.figure(figsize=(10, 5))
dendrogram(linked)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Students')
plt.ylabel('Distance')
plt.show()
```
