# CLO2 Analysis – Social Media Addiction Dataset
This notebook covers: descriptive statistics, sampling, visualizations, correlation tests, chi-square, and one-sample t-test.

## 1. Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## 2. Load dataset

```python
df = pd.read_csv("Students Social Media Addiction.csv")

df.head()
```

```python
print("Dataset shape (rows, cols):", df.shape)
df.head()
```

## 3. Data overview

```python
df.info()
df.describe()
```

```python
df.describe()
```

## 4. Descriptive statistics function (Dependent Variable: Addicted_Score)

```python
def descriptive_statistics(data):

    mean = data["Addicted_Score"].mean()
    median = data["Addicted_Score"].median()
    std = data["Addicted_Score"].std()
    minimum = data["Addicted_Score"].min()
    maximum = data["Addicted_Score"].max()

    print("Addicted Score Statistics")
    print("------------------------")
    print("Mean:", mean)
    print("Median:", median)
    print("Standard Deviation:", std)
    print("Minimum:", minimum)
    print("Maximum:", maximum)
```

```python
descriptive_statistics(df)
```

## 5. Random sampling (n = 150)

```python
random_sample = df.sample(n=150, random_state=42)

random_sample.head()
```

```python
print("Random sample size:", len(random_sample))
random_sample.head()
```

```python
descriptive_statistics(random_sample)
```

## 6. Systematic sampling (~150 using interval k)

```python
k = len(df) // 150
systematic_sample = df.iloc[::k].copy()
print("k (interval):", k)
print("Systematic sample size:", len(systematic_sample))
systematic_sample.head()
```

```python
descriptive_statistics(systematic_sample)
```

## 7. Visualizations for Addicted_Score

### 7.1 Scatter plot: Avg_Daily_Usage_Hours vs Addicted_Score

```python
plt.scatter(df["Avg_Daily_Usage_Hours"], df["Addicted_Score"])

plt.xlabel("Daily Usage Hours")
plt.ylabel("Addicted Score")
plt.title("Relationship Between Social Media Usage and Addiction")

plt.show()
```

```python
sns.boxplot(x=df["Addicted_Score"])

plt.title("Boxplot of Addiction Score")

plt.show()
```

### 7.3 Histogram: Addicted_Score

```python
plt.hist(df["Addicted_Score"], bins=20)

plt.title("Distribution of Addiction Score")
plt.xlabel("Addicted Score")
plt.ylabel("Frequency")

plt.show()
```

### 7.4 Heatmap: Correlation (numeric variables only)

```python
corr = df.drop(columns=["Student_ID"]).corr(numeric_only=True)

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

## 8. Hypothesis testing

### 8.1 Pearson correlation (numeric IV vs DV)

```python
pearson_corr, p_value = stats.pearsonr(df["Avg_Daily_Usage_Hours"], df["Addicted_Score"])

print("Pearson Correlation:", pearson_corr)
print("P-value:", p_value)
```

### 8.2 Spearman correlation (numeric IV vs DV)

```python
spearman_corr, p_value = stats.spearmanr(df["Sleep_Hours_Per_Night"], df["Addicted_Score"])

print("Spearman Correlation:", spearman_corr)
print("P-value:", p_value)
```

### 8.3 Chi-square test (categorical vs categorical)

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(df["Gender"], df["Relationship_Status"])

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square:", chi2)
print("P-value:", p)
```

## 9. One-sample t-test (Is the random sample representative of the dataset?)

```python
population_mean = df["Addicted_Score"].mean()
t_stat, p_value = stats.ttest_1samp(random_sample["Addicted_Score"], population_mean)
print("Population mean (Addicted_Score):", population_mean)
print("Random sample mean:", random_sample["Addicted_Score"].mean())
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
print(X)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in range(X.shape[1]):
    if isinstance(X[0, i], str):
        X[:, i] = le.fit_transform(X[:, i])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
